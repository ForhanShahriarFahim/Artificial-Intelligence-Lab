```python
import numpy as np
import matplotlib.pyplot as plt


# Linear function: y = ax + b
def linear_function(x, a, b):
    return a * x + b


# Nonlinear function: y = a * sin(x) + b
def sinusoidal_function(x, a, b):
    return a * np.sin(x) + b


# Nonlinear function: y = a * e^x + b
def exponential_function(x, a, b):
    return a * np.exp(x) + b


# Nonlinear function: y = ax^2 + b
def quadratic_function(x, a, b):
    return a * x**2 + b


# Generate x values
x_values = np.linspace(-10, 10, 100)

# Parameters for the functions
a = 1
b = 2

# Linear function
y_linear = linear_function(x_values, a, b)

# Sinusoidal function
y_sinusoidal = sinusoidal_function(x_values, a, b)

# Exponential function
y_exponential = exponential_function(x_values, a, b)

# Quadratic function
y_quadratic = quadratic_function(x_values, a, b)

# Plotting the functions
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x_values, y_linear, label="Linear Function",marker='o')
plt.title("Linear Function")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x_values, y_sinusoidal, label="Sinusoidal Function", marker="o")
plt.title("Sinusoidal Function")
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x_values, y_exponential, label="Exponential Function", marker="o")
plt.title("Exponential Function")
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x_values, y_quadratic, label="Quadratic Function", marker="o")
plt.title("Quadratic Function")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Linear function: y = ax + b
def linear_function(x, a, b):
    return a * x + b


# Nonlinear function: y = a * sin(x) + b
def sinusoidal_function(x, a, b):
    return a * np.sin(x) + b


# Nonlinear function: y = a * e^x + b
def exponential_function(x, a, b):
    return a * np.exp(x) + b


# Nonlinear function: y = ax^2 + b
def quadratic_function(x, a, b):
    return a * x**2 + b


# Generate x values
x_values = np.linspace(-10, 10, 10)

# Parameters for the functions
a = 1
b = 2

# Linear function
y_linear = linear_function(x_values, a, b)

# Sinusoidal function
y_sinusoidal = sinusoidal_function(x_values, a, b)

# Exponential function
y_exponential = exponential_function(x_values, a, b)

# Quadratic function
y_quadratic = quadratic_function(x_values, a, b)

# Plotting the functions with markers and annotations
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x_values, y_linear, label="Linear Function", marker="o")
for x, y in zip(x_values, y_linear):
    plt.annotate(f"({x:.2f}, {y:.2f})", (x, y))
plt.title("Linear Function")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x_values, y_sinusoidal, label="Sinusoidal Function", marker="x")
for x, y in zip(x_values, y_sinusoidal):
    plt.annotate(f"({x:.2f}, {y:.2f})", (x, y))
plt.title("Sinusoidal Function")
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x_values, y_exponential, label="Exponential Function", marker="s")
for x, y in zip(x_values, y_exponential):
    plt.annotate(f"({x:.2f}, {y:.2f})", (x, y))
plt.title("Exponential Function")
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x_values, y_quadratic, label="Quadratic Function", marker="^")
for x, y in zip(x_values, y_quadratic):
    plt.annotate(f"({x:.2f}, {y:.2f})", (x, y))
plt.title("Quadratic Function")
plt.legend()

plt.tight_layout()
plt.show()
```
