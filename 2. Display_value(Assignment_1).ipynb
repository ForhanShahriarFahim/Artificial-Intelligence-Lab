# Importing Library

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Function 1: y = wx + b

```python
def linear_function(x, w, b):
  return w * x + b
```

### Function 2: y = x^2

```python
def quadratic_function(x):
  return x**2
```

### Function 3: y = 1 / (1 + e^(-x))

```python
def sig_function(x):
  return 1 / (1 + np.exp(-x))
```

### Function 4: y = (e^x - e^(-x)) / (e^x + e^(-x))

```python
def tanh_function(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

### Function 5: g(f(x)) where u = f(x) = wx + b and y = g(u) = 1 / (1 + e^(-x))

```python
def composite_5(x, w, b):
  u = linear_function(x, w, b)
  return sig_function(u)
```

### Function 6: g(f(x)) where f(x) = wx+b, u = f(x) and g(u) = (e^u - e^(-u)) / (e^u - e^(-u))

```python
def composite_6(x, w, b):
  u = linear_function(x, w, b)
  return tanh_function(u)
```

```python
# Function 7: y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x) and g1(u1) = 1 / (1 + e^(-u1))
# y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x) and g2(u2) = 1 / (1 + e^(-u2))
# y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v) and g3(w) = 1 / (1 + e^(-w))

def composite_7(x, w1, b1, w2, b2, w3, w4, b):
  u1 = linear_function(x, w1, b1)
  u2 = linear_function(x, w2, b2)
  y1 = sig_function(u1)
  y2 = sig_function(u2)

  v = w3 * y1 + w4 * y2 + b
  return sig_function(v)
```

```python
# Range of x values
x_values = np.arange(-10, 10.1, 0.1)
```

```python
# Define constants
w_linear = 2
b_linear = 3

w_sigmoid_on_linear = 0.5
b_sigmoid_on_linear = -1

w_tanh_on_linear = -1
b_tanh_on_linear = 1

w1 = 0.5
b1 = -1
w2 = -0.5
b2 = 1
w3 = 2
w4 = -1
b = 0.5
```

```python
# Calculation values for each function
y_linear = linear_function(x_values, w_linear, b_linear)
y_quadratic = quadratic_function(x_values)
y_sigmoid = sig_function(x_values)
y_tanh = tanh_function(x_values)
y_composite_fun_5 = composite_5(x_values, w_sigmoid_on_linear, b_sigmoid_on_linear)
y_composite_fun_6 = composite_6(x_values, w_tanh_on_linear, b_tanh_on_linear)
y_composite_fun_7 = composite_7(x_values, w1, b1, w2, b2, w3, w4, b)
```

```python
# Plotting the output
plt.figure(figsize=(12,10))

# Linear Function
plt.subplot(4,2,1)
plt.plot(x_values, y_linear, label = 'Linear Function')
plt.title("Linear Function")
plt.legend()

# Quadratic Function
plt.subplot(4,2,2)
plt.plot(x_values, y_quadratic, label = 'Quadratic Function')
plt.title('Quadratic Function')
plt.legend()

# Sigmoid Function
plt.subplot(4,2,3)
plt.plot(x_values, y_sigmoid, label = 'Sigmoid Function')
plt.title('Sigmoid Function')
plt.legend()

# Tanh Function
plt.subplot(4,2,4)
plt.plot(x_values, y_tanh, label = 'Tanh Function')
plt.title('Tanh Function')
plt.legend()

# Composite Function 5
plt.subplot(4,2,5)
plt.plot(x_values, y_composite_fun_5, label = 'Comp Func-5')
plt.title('Composite-5')
plt.legend()

# Composite Function 6
plt.subplot(4,2,6)
plt.plot(x_values, y_composite_fun_6, label = 'Comp Func-6')
plt.title('Composite-6')
plt.legend()

# Composite Functon 7
plt.subplot(4,2,7)
plt.plot(x_values, y_composite_fun_7, label = 'Comp Func-7')
plt.title('Composite-7')
plt.legend()

plt.tight_layout()
plt.show()
```
