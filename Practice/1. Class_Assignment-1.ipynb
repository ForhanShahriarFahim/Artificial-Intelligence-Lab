```python
import matplotlib.pyplot as plt 
import numpy as np

def linear_function(x, a, b):
    return a*x+ b

def sinusoidal_function(x, a, b):
    return a * np.sin(x) + b

def exponential_function(x,a,b):
    return a * np.exp(x) + b

def quadratic_function(x, a, b): 
    return a*x**2+b

x_values = np.linspace(-10, 10, 400)

a, b = 1,2

y_linear = linear_function(x_values, a,b)

y_sinusoidal = sinusoidal_function(x_values, a,b)

y_exponential = exponential_function(x_values,a,b)

y_quadratic = quadratic_function(x_values, a,b)

# plotting the functions 
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.plot(x_values, y_linear, label = "Linear Function")
plt.title("Linear Function")
plt.legend()


plt.subplot(2,2,2)
plt.plot(x_values, y_sinusoidal, label = "Sinusoidal Function")
plt.title("sinusoidal Function")
plt.legend()

plt.subplot(2,2,3)
plt.plot(x_values, y_exponential, label = "Exponential Function")
plt.title("Exponential Function")
plt.legend()

plt.subplot(2,2,4)
plt.plot(x_values, y_quadratic, label = 'Quadratic Function')
plt.title('Quadratic Function')
plt.legend()

plt.tight_layout()
plt.show()
```
